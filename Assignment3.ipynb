```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("student_info.csv")
df = df.drop_duplicates()
df = df.dropna() 
df['gender'] = df['gender'].str.strip().str.capitalize()
df['parent_education'] = df['parent_education'].str.strip().str.title()
df['internet_access'] = df['internet_access'].str.strip().str.capitalize()
df['lunch_type'] = df['lunch_type'].str.strip().str.title()
df['extra_activities'] = df['extra_activities'].str.strip().str.capitalize()
df['final_result'] = df['final_result'].str.strip().str.capitalize()
plt.hist(df['study_hours'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Study Hours')
plt.xlabel('Hours')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',
        colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(df['math_score'], kde=True, color='skyblue')
plt.title('Math Score')
plt.subplot(1, 3, 2)
sns.histplot(df['reading_score'], kde=True, color='lightgreen')
plt.title('Reading Score')
plt.subplot(1, 3, 3)
sns.histplot(df['writing_score'], kde=True, color='salmon')
plt.title('Writing Score')
plt.tight_layout()
plt.show()
avg_scores = df[['math_score', 'reading_score', 'writing_score']].mean()
plt.bar(avg_scores.index, avg_scores.values, color=['blue', 'green', 'red'])
plt.title('Average Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(True, axis='y')
plt.show()
result_counts = df['final_result'].value_counts()
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%',
        colors=['lightblue', 'lightgray'])
plt.title('Final Result Distribution')
plt.axis('equal')
plt.show()
edu_counts = df['parent_education'].value_counts()
plt.bar(edu_counts.index, edu_counts.values, color='purple')
plt.title('Parent Education Level')
plt.xticks(rotation=30)
plt.xlabel('Education Level')
plt.ylabel('Number of Students')
plt.grid(True, axis='y')
plt.show()
plt.scatter(df['study_hours'], df['math_score'], alpha=0.7, color='teal')
plt.title('Study Hours vs Math Score')
plt.xlabel('Study Hours')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()
color_map = {'Pass': 'green', 'Fail': 'red'}
colors = df['final_result'].map(color_map)
plt.scatter(df['attendance_rate'], df['study_hours'], c=colors, alpha=0.6)
plt.title('Attendance Rate vs Study Hours\n(Colored by Final Result)')
plt.xlabel('Attendance Rate (%)')
plt.ylabel('Study Hours')
plt.grid(True)
plt.show()
sns.pairplot(df[['math_score', 'reading_score', 'writing_score', 'final_result']], hue='final_result', palette='Set1')
plt.suptitle("Pairwise Relationships of Scores", y=1.02)
plt.show()
corr = df[['math_score', 'reading_score', 'writing_score', 'attendance_rate', 'study_hours', 'age']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
